```python
# Now..
cols_dict=dict(Visa_dataset.dtypes)
cat_list=[]
for key,value in cols_dict.items():
    if value=='O':
        cat_list.append(key)
        
cat_list
```

```python
# If we want to add both the lists, i.e. catagorical and numerical list, then...

cols_dict=dict(Visa_dataset.dtypes)
cat_list,num_list=[],[]
for key,value in cols_dict.items():
    if value=='O':
        cat_list.append(key)
    else:
        num_list.append(key)

cat_list,num_list
```

```python
# Now, doing the above in "List-Comprehension"...

cat_list=[key for key,value in cols_dict.items() if value=='O']

num_list=[key for key,value in cols_dict.items() if value!='O']

cat_list,num_list
```
